## Clustering (Crime data)

```python
Perform Clustering(Hierarchical, Kmeans & DBSCAN) for the crime data and identify the number of clusters formed and draw inferences.

Data Description:
Murder -- Muder rates in different places of United States
Assualt- Assualt rate in different places of United States
UrbanPop - urban population in different places of United States
Rape - Rape rate in different places of United States
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
crim=pd.read_csv('C:\\Users\\Hp\\crime_data.csv')

crim
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)

df_norm=norm_func(crim.iloc[:,1:])

df_norm
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
# create clusters
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(df_norm)
y_hc
```

```python
crim['h_clusterid']=hc.labels_
crim
```

###### Kmeans

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist


crim1=pd.read_csv('C:\\Users\\Hp\\crime_data.csv')
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)


df_norm=norm_func(crim.iloc[:,1:])
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crim1['Clust']=x


crim1
```

```python
crim1.iloc[:,1:5].groupby(crim1.Clust).mean()
```

###### DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler


crim3=pd.read_csv('C:\\Users\\Hp\\crime_data.csv')


crim3.info()
```

```python
df=crim3.iloc[:,1:5]
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)


x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([crim3,cl],axis=1)
```
